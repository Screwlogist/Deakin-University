2. Fill in the missing entries. For filling any feature, you can use either the mean or median value of the feature values from observed entries. Explain the reason behind your choice and print replacement value of each feature.

Answer: Since the skewness of the feature "ELEVATION" is 0.5 that is consider has symmetrical

```python
#Finding skewness of the data
df.skew()
```

```python
#Copying data to a different dataframe
df1 = df.copy()
```

```python
#Filling all NA values with 0 to find the mean.
df1 = df1.fillna(0)
df1
```

```python
#Filtering location with the name DOCKLANDS LIBRARY, Since we know the missing values from elevation are releated to DOCKLANDS LIBRARY location.
loc_dock = df1[df['location'] == "Docklands Library"]
loc_dock
```

```python
#Importing mean from Statustics and finding the mean value and storing the mean value
from statistics import mean

elevation_mean = mean(loc_dock['elevation'])
elevation_mean
```

```python
df3 = df.copy()
df3
```

```python
#Replacing NA values with the newly found mean value 
df3['elevation'] = df3['elevation'].fillna(elevation_mean)
df3
```

```python
#Verifying any further missing values. We are ignoring Position since it is not related to any other feature in the provided dataset.
df3.isna().sum()
```

4. Apply the min-max scaling on features "[boardtype, boardid ,temp_max, temp_min, temp_avg,light_max, light_min,light_avg, humidity_min, humidity_max, humidity_avg]" . Plot distribution of first six features before and after scaling. Is there any difference? Please explain.

```python
#Finding the datatypes of all the features since Min-Max Scaling works only for numerical datatypes
df.dtypes
```

```python
#Storing the features in a variable
X_features = df.iloc[:,2:13]
X_features
```

```python
#Importing MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
#Fitting the object into the MinMaxScaler and storing into an Object
scaler = MinMaxScaler().fit(X_features)
```

```python
#Finding the min and max value for each feature
X_features.describe()
```

```python
#Verifying the MinMaxScaler range
scaler.feature_range
```

We can verify the Min-Max Scaling by using the formula
(Value - Min) / (Max - Min)

```python
#Applying the MinMaxScaler Normalization
X_features_scaled = scaler.transform(X_features)
X_feature_scaled_df = pd.DataFrame(X_features_scaled, columns=X_features.columns)
```

```python
#Verifying after MinMaxScaling
X_feature_scaled_df.describe()
```
